```r
bar <- ggplot(data = diamonds) + 
       geom_bar(mapping = aes(x = cut, y=stat(count/1000), fill = color),
                show.legend = TRUE,
                width = 0.8) + 
       theme(aspect.ratio = 1) + 
       labs(x = NULL, y = NULL)+
       facet_wrap(facets = vars(clarity))
```
