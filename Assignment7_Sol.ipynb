```python
diamonds = pd.read_csv('data/diamonds.csv')
prices = np.array(diamonds['price'])
carats = np.array(diamonds['carat'])
```

### Q1 (2 points): Print the mean, standard deviation, minimum, maximum, 25th percentile, median, and 75th percentile of the prices array.

```python
mean_price = np.mean(prices)
standard_devation = np.std(prices)
minimum_price = np.min(prices)
maximum_price = np.max(prices)
percentile_25th = np.percentile(prices, 25) 
median = np.median(prices)
percentile_75th = np.percentile(prices, 75) 

print("========================================")
print("Mean price:        " + str(mean_price))
print("Standard deviation: " + str(standard_devation))
print("Minimum price:     " + str(minimum_price))
print("Maximum price:     " + str(maximum_price))
print("========================================")
print("25th percentile:    " + str(percentile_25th))
print("Median:             " + str(median))
print("75th percentile:    " + str(percentile_75th))
print("========================================")
```

### Q2 (2 points): Plot the histogram of the prices array.

```python
plt.hist(prices)
plt.title('Price Distribution')
plt.xlabel('prices')
plt.ylabel('number');
```
